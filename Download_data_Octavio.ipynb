```python
import requests
import pandas as pd
from urllib import request
```

```python
#Se consulta la información en la URL Solicitada
try:
    url = "https://www.cpc.ncep.noaa.gov/products/precip/CWlink/pna/norm.nao.monthly.b5001.current.ascii.table"
    file_url = url
    r = requests.get(url)
    print ("DATA:\n", r.text)

except:
    print ("URL Invalida, algún error ocurrió al realizar la consulta a la URL solicitada")
    
```

```python
def download_file_information(url):
    fileOpen =request.urlopen(url)
    file_info = fileOpen.read()
    file_info_str = str(file_info)
    file_lines = file_info_str.split('\\n')
    data_file =open ('fiel.txt', "w")
    for info in file_lines:
        data_file.write(info +'\n')
        
    data_file.close()
download_file_information(file_url)
```
